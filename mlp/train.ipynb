```python
import pandas as pd
import torch

import mlp_scratch
import mlp_torch_nn
```

```python
dataset = pd.read_csv("data/IRIS_flower.csv").values
```

```python
# Create dataset for training
xs = torch.tensor([list(x[:-1]) for x in dataset], dtype=torch.float32)
ys_raw = [x[-1] for x in dataset]

# Create a dictionary from y -> idx & vice-versa
idx_to_ys = {}
for idx, name in enumerate(set(ys_raw)):
    idx_to_ys[idx] = name

ys_to_idx = {}
for idx, val in idx_to_ys.items():
    ys_to_idx[val] = idx

# One-Hot Encode Ys
# format ys to idx
ys = torch.zeros(150, 3)
for i in range(len(ys_raw)):
    ys[i][ys_to_idx[ys_raw[i]]] = 1
```

```python
# Define both MLPs with input dim: 4 
# and two hidden layers (16 neurons each)
# and 3 output neurons

mlp_raw = mlp_scratch.MLP(4, [16, 16, 3])
mlp_nn = mlp_torch_nn.MLP(4, [16, 16, 3])
```

```python
mlp_raw.train(xs, ys, lr=0.05, epochs=500)
```

```python
mlp_raw.plot_model(metric='loss')
```

```python
mlp_nn.train(xs, ys, epochs=500)
```

```python
mlp_nn.plot_model('loss')
```
